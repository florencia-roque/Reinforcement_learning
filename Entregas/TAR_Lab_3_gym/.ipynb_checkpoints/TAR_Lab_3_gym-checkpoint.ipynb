# Definici√≥n de las pol√≠ticas de decisi√≥n

Como Q-learning es un algoritmo off-policy, se utiliza una pol√≠tica de decisi√≥n para actuar y otra diferente para actualizar la tabla Q.
Se utilizar√°n:
- Pol√≠tica greedy (pol√≠tica de actualizaci√≥n)
- Pol√≠tica epsilon-greedy (pol√≠tica de actuaci√≥n)

#### 1.2 Crear una funci√≥n que implemente la pol√≠tica greedy. Las entradas son la tabla Q y el estado actual, y la salida es la acci√≥n a tomar.

````python
#TODO crear una funcion que implemente la politica greedy.

def greedy_policy(Q, state):
    """
    Pol√≠tica greedy: devuelve la acci√≥n con mayor valor Q para un estado dado.
    
    Par√°metros:
    - Q: tabla Q de tama√±o [n_states, n_actions]
    - state: estado actual (int)

    Retorna:
    - acci√≥n (int) con el mayor valor Q
    """
    return np.argmax(Q[state])
````

#### 1.3 Crear otra funci√≥n que implemente la pol√≠tica epsilon-greedy. 
Adem√°s de los par√°metros de entrada de la funci√≥n anterior, agregar $œµ$, que es la variable que controla la relaci√≥n exploraci√≥n/explotaci√≥n a la hora de decidir que acci√≥n tomar.

````python
#TODO crear una funcion que implemente la politica epsilon greedy.

def epsilon_greedy_policy(Q, state, epsilon):
    """
    Pol√≠tica epsilon-greedy: explora con probabilidad epsilon, explota (greedy) con 1-epsilon.
    
    Par√°metros:
    - Q: tabla Q de tama√±o [n_states, n_actions]
    - state: estado actual (int)
    - epsilon: probabilidad de explorar (float)
    
    Retorna:
    - acci√≥n (int)
    """
    n_actions = Q.shape[1]
    
    # 1 con probabilidad epsilon, 0 con probabilidad 1-epsilon
    explorar = np.random.binomial(1,epsilon)

    if explorar == 1:
        # exploro, sorteo con distribucion uniforme 
        action = np.random.choice(n_actions)
    else:
        # exploto 
        action = greedy_policy(Q,state)

    return action
````

Una vez creadas las funciones anteriores, solamente falta crear la funci√≥n de entrenamiento. La misma debe seguir el siguiente pseudo-c√≥digo:

```
Para cada episodio en el total de episodios de entrenamiento:

Reducir epsilon (ya que necesitamos cada vez menos exploraci√≥n)
Reiniciar el entorno

  Para cada paso en el m√°ximo n√∫mero de timesteps:
    Elegir la acci√≥n At utilizando la pol√≠tica epsilon-greedy
    Realizar la acci√≥n (a) y observar el nuevo estado (s') y la recompensa (r)
    Actualizar el valor de Q(s,a) usando la ecuaci√≥n de Bellman
    Si el episodio ha terminado, finalizarlo

```

**Actualizaci√≥n de la Tabla Q**
  
La funci√≥n de valor Q se actualiza utilizando la ecuaci√≥n de Bellman:  
  
$$ Q(s, a) \leftarrow Q(s, a) + \alpha \left[ R + \gamma \max_{a'} Q(s', a') - Q(s, a) \right] $$
  
donde:  
- $ \alpha $ es la tasa de aprendizaje.  
- $ \gamma $ es el factor de descuento.  
- $ R $ es la recompensa recibida.  
- $ s' $ es el nuevo estado.  
- $ a' $ es la acci√≥n que maximiza el valor Q en el nuevo estado.

**Observaci√≥n**: Es importante que la variable `epsilon` varie entre un valor m√°ximo y m√≠nimo, y que vaya decayendo de manera exponencial seg√∫n el `decay_rate` y el n√∫mero de episodio en el que se est√©.

**Observaci√≥n 2**: Almacenar la ganacia acumulada para cada episodio para poder graficar la evoluci√≥n de entrenamiento

#### 1.4 Crear la funci√≥n de entrenamiento

````python
def train(n_training_episodes, min_epsilon, max_epsilon, decay_rate, env, max_steps, gamma, learning_rate, Qtable):
    rewards = []

    for episode in range(n_training_episodes):
        
        if episode < 5000:
            epsilon = 1.0  # explorar completamente
        else:
            epsilon = min_epsilon + (max_epsilon - min_epsilon) * np.exp(-decay_rate * (episode - 5000))

        
        # # Decaimiento exponencial de epsilon
        # epsilon = min_epsilon + (max_epsilon - min_epsilon) * np.exp(-decay_rate * episode)

        state, _ = env.reset()
        done = False
        total_rewards = 0

        for step in range(max_steps):
            # Acci√≥n con pol√≠tica epsilon-greedy
            action = epsilon_greedy_policy(Qtable, state, epsilon)

            new_state, reward, terminated, truncated, info = env.step(action)
            done = terminated or truncated

            # Actualizar Q usando la ecuaci√≥n de Bellman
            best_next_action_q = np.max(Qtable[new_state])  
            Qtable[state, action] = Qtable[state, action] + learning_rate * (reward + gamma * best_next_action_q - Qtable[state, action])

            # Acumular recompensa y avanzar de estado
            total_rewards += reward
            state = new_state

            if done:
                break

        rewards.append(total_rewards)

    return Qtable, rewards
````

#### 1.5 Definir los hiperpar√°metros, inicializar la tabla Q llena de ceros y entrenar el agente

````python
# Par√°metros de entrenamiento
n_training_episodes = 10000  # Total de episodios de entrenamiento
learning_rate = 0.1          # Tasa de aprendizaje

# Par√°metros del entorno
max_steps = 200              # M√°ximo de pasos por episodio
gamma = 0.99                 # Tasa de descuento

# Par√°metros de exploraci√≥n
max_epsilon = 1.0           # Probabilidad de exploraci√≥n al inicio
min_epsilon = 0.01           # Probabilidad m√≠nima de exploraci√≥n
decay_rate = 0.0001         # Tasa de decaimiento de epsilon

# inicializar la tabla Q con ceros
Qtable_frozenlake = np.zeros((64,4))
````

````python
np.random.seed(42)
Qtable_frozenlake, rewards = train(n_training_episodes, min_epsilon, max_epsilon, decay_rate, env, max_steps, gamma, learning_rate, Qtable_frozenlake)
````

#### 1.6 El siguente c√≥digo grafica el promedio de recompensas por cada bloque de 1000 episodios. Observar y explicar su comportamiento.

````python
# Calculamos el promedio de recompensas por cada bloque de 1000 episodios
reward_per_thousand_episodes = np.split(np.array(rewards), n_training_episodes/1000)
count = 1000
avg_rewards = []

for r in reward_per_thousand_episodes:
    avg_rewards.append(sum(r/1000))
    count += 1000

# Visualizamos el promedio de recompensas
plt.plot(avg_rewards)
plt.xlabel('Bloques de 1000 episodios')
plt.ylabel('Promedio de Recompensas')
plt.title('Promedio de Recompensas por Episodio durante el Entrenamiento')
plt.grid()
plt.show()
````

````python
def print_policy(Q, env):
    symbols = ['‚Üê', '‚Üì', '‚Üí', '‚Üë']
    policy = np.argmax(Q, axis=1)
    shape = env.desc.shape
    grid = policy.reshape(shape)

    print("Pol√≠tica greedy aprendida:")
    for fila in grid:
        print(' '.join(symbols[a] for a in fila))
````

#### 1.7 Simular un juego y obtener la ganancia (reward) total acumulada. Ver si el agente logr√≥ llegar a la meta (si el roward total es positivo).

En caso contrario variar los hiperpar√°metros.


````python
total_reward = 0
state, _ = env.reset()
done = False

# Jugar una vez con la tabla Q entrenada y obtener la ganancia acumulada
for step in range(max_steps): 
    action = greedy_policy(Qtable_frozenlake, state)
    new_state, reward, terminated, truncated, _ = env.step(action)
    done = terminated or truncated

    total_reward += reward
    state = new_state

    if done:
        break

if total_reward > 0:
  print('Llegaste a la meta!')
print('total_reward ', total_reward)
````

#### 1.8 Qu√© impacto tiene en la soluci√≥n obtenida los diferentes hiperpar√°metros? Qu√© sucede al variar el `decay_rate` y `max_epsilon` y `min_epsilon`?

##### RESPUESTA -->

El max_epsilon condiciona la probabilidad de exploraci√≥n inicial, un valor cercano a 1 indica que el agente explora mucho al principio, lo que es razonable porque al principio el agente no tiene idea donde est√° el objetivo ni qu√© acciones son peligrosas.

El min_epsilon es el valor al que ir√≠a convergiendo el epsilon a medida que pasan los episodios, un valor bajo (por ejemplo cercano a 0.01) sirve porque as√≠ el agente termina casi completamente explotando, esto es bueno si el agente ya aprendi√≥. Si se toma un valor alto, por ejemplo cercano a 0.5, puede hacer que incluso con una buena pol√≠tica siga eligiendo acciones al azar y nunca deje de explorar, esto baja el rendimiento del agente. En entornos estoc√°sticos, donde la siguiente acci√≥n tiene una distribuci√≥n de probabilidad asociada puede ser que convenga tener un min_epsilon un poco m√°s alto. 

El decay_rate representa la velocidad con la que epsilon decae. Si es muy alto, epsilon cae muy r√°pido, el agente deja de explorar antes de haber aprendido algo √∫til. Si es muy bajo, el agente sigue explorando aunque ya sabe que hacer, puede tardar en converger.

#### 1.9 Si se quisiera un agente que tome el camino m√°s corto, ¬øqu√© soluci√≥n/es se le ocurre? Implementarla/s

##### RESPUESTA -->

Q-learning no garantiza el camino m√°s corto, porque el agente maximiza recompensa acumulada y solo recibe 1 al llegar a la meta y 0 en los pasos intermedios, entonces todos los caminos v√°lidos tienen la misma recompensa. Eso significa que el agente no tiene incentivo para elegir el camino m√°s corto.

**Alternativa:**

Modificar la funci√≥n de recompensa:

reward = -1 en cada paso

reward = 0 al llegar al objetivo

Esto hace que:

Cuantos m√°s pasos tome, m√°s recompensa negativa acumula.

Entonces, el agente podr√≠a aprender a minimizar la cantidad de pasos.

````python
from gymnasium import RewardWrapper

class ShortestPathRewardWrapper(RewardWrapper):
    def reward(self, reward):
        # Penaliza cada paso con -1
        return reward - 1

env = gym.make("FrozenLake-v1", map_name="8x8", is_slippery=False)
env = ShortestPathRewardWrapper(env)
````

````python
np.random.seed(42)
Qtable_frozenlake, rewards = train(n_training_episodes, min_epsilon, max_epsilon, decay_rate, env, max_steps, gamma, learning_rate, Qtable_frozenlake)
````

````python
# Calculamos el promedio de recompensas por cada bloque de 1000 episodios
reward_per_thousand_episodes = np.split(np.array(rewards), n_training_episodes/1000)
count = 1000
avg_rewards = []

for r in reward_per_thousand_episodes:
    avg_rewards.append(sum(r/1000))
    count += 1000

# Visualizamos el promedio de recompensas
plt.plot(avg_rewards)
plt.xlabel('Bloques de 1000 episodios')
plt.ylabel('Promedio de Recompensas')
plt.title('Promedio de Recompensas por Episodio durante el Entrenamiento')
plt.grid()
plt.show()
````

**Observaciones:** 
Al modificar la funci√≥n de recompensa penalizando los pasos (reward - 1), el agente aprende a minimizar la cantidad de pasos hacia la meta, no solo a llegar.

La siguente funci√≥n toma la tabla Q entrenada y simula un juego, guardando los frames y generando un video.

````python
from matplotlib import animation
def generate_frames(q_table, env, max_steps=100):
    """Genera una lista de frames para crear la animaci√≥n"""
    state, _ = env.reset()
    frames = [env.render()]  # Render del estado inicial
    for step in range(max_steps):
        action = greedy_policy(q_table, state)
        next_state, reward, done, _, _ = env.step(action)
        frames.append(env.render())  # Guardar el nuevo estado como frame

        state = next_state

        if done:
            break
    return frames

def save_video(frames, filename='q_learning_frozenlake.mp4'):
    """Guardar el video de los frames en un archivo .mp4"""
    fig = plt.figure()
    patch = plt.imshow(frames[0])
    plt.axis('off')

    def animate(i):
        patch.set_data(frames[i])

    anim = animation.FuncAnimation(fig, animate, frames=len(frames), interval=300)

    # Guardar animaci√≥n
    anim.save(filename, writer='ffmpeg', fps=5)
    plt.close()
````
